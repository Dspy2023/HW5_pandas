## 題目一 (50%)

這題會使用到鳶尾花資料集 （Iris dataset），這個資料集中包括含四種花卉屬性與三種花卉種類，分別是：
- 屬性
  - Sepal Length: 花萼長度 (cm)
  - Sepal Width: 花萼寬度 (cm)
  - Petal Length: 花瓣長度 (cm)
  - Petal Width: 花瓣寬度 (cm)
- 種類
  - Setosa: 山鳶尾
  - Versicolor: 變色鳶尾
  - Virginica: 維吉尼亞鳶尾

### 程式要求

1. 讀取資料
  - 讀取 `iris.csv` 資料集，並將其存放在 `iris_df`。
  - 使用 `.head()` 觀察一下 `iris_df` 的數據（不用印出來）。

2. 前處理
  - 移除有缺失值（NaN）的資料，並重新編排index（用`.reset_index()`）。
  - 刪除名為 'Id' 的欄位。
  - 將剩餘的欄位重新命名，方式為將屬性與單位用底線隔開，例如'SepalLength_Cm'，若無單位則維持原本的欄位名。
  - 將更新後的dataframe用 `.head()` 印出來。

3. 新增欄位
  - 計算 'SepalLength_Cm' 與 'SepalWidth_Cm' 欄位的比率，將結果儲存在 'SepalRatio' 欄位。
  - 計算 'PetalLength_Cm' 與 'PetalWidth_Cm' 欄位的乘積，將結果儲存在 'PetalArea' 欄位。
  - 將更新後的dataframe用 `.head()` 印出來。

4. 資料索引
  - 使用 `.iloc()` ，選取資料集中第 6 到 11 個橫列（含第11）以及所有的直欄，將結果存放在 `subset_iloc` 中，然後全部印出來。
  - 使用 `.loc()` ，選取資料集中標籤為 10 到 20 的橫列，以及標籤為'SepalWidth_Cm'和'PetalWidth_Cm'的直欄，將結果存放在 `subset_loc` 中，然後全部印出來。


5. 資料篩選
  - 從資料集中篩選出 'Species' 為 "Iris-versicolor" 的資料，將結果存放在 `versicolor_df` 中，然後用 `.head()` 印出來。
  - 在 `versicolor_df` 中進一步篩選出 'SepalLength_Cm' 大於 6.0 且 'PetalLength_Cm' 大於 4.0 的資料，並將結果存放在 `filtered_df` 中，然後全部印出來。
  - 計算符合條件的資料筆數，並將結果存放在count中，然後印出來。

6. 函數
  - 定義一個名為 `summary_stats() `的函數，它接受兩個參數： `dataframe` 和 `column_name`。
  - 這個函數會計算並回傳三個統計值：
    - 'mean'：欄位的平均值。
    - 'median'：欄位的中位數。
    - 'std'：欄位的標準差。
  - 呼叫函數，計算'SepalLength_Cm'欄位的統計值，將結果存放在 sepal_length_summary 中，然後印出來。

這是一個關於現代奧運會的歷史資料集，包含從1896年雅典到2016年里約的所有賽事。[資料來源](https://www.kaggle.com/datasets/heesoo37/120-years-of-olympic-history-athletes-and-results)

> **請注意**：直到1992年，夏季和冬季運動會是在同一年舉行的。之後它們就分開，冬季運動會在1994年開始，然後夏季運動會在1996年，再然後是冬季運動會在1998年，以此類推。分析這些數據時，人們常犯的一個錯誤是假設夏季和冬季運動會一直都是分開的。

### 內容
資料集 `athlete_events.csv` 包含271116個橫列和15個直欄。每行對應一位運動員參加一個奧運賽事。這些列包括：

- ID - 每個運動員的唯一編號
- Name - 運動員的名字
- Sex - 男（M）或女（F）
- Age - 年齡（整數）
- Height - 身高（以厘米為單位）
- Weight - 體重（以千克為單位）
- Team - 隊伍名稱
- NOC - 國家奧林匹克委員會的3個字母代碼
- Games - 奧運年份和賽季
- Year - 奧運年份（整數）
- Season - 奧運賽季（夏季或冬季）
- City - 主辦城市
- Sport - 運動項目
- Event - 賽事
- Medal - 獲得的獎牌（金牌、銀牌、銅牌或NA）
